# üè® Hotel Booking Data Analysis & Insights

Analyzing hotel booking data to uncover insights into customer behavior, booking patterns, and factors influencing cancellations. This project uses Python libraries like pandas, seaborn, and matplotlib to visualize and explore the data.

## üéØ Project Objectives
- Understand customer booking behavior
- Analyze cancellation trends
- Explore the impact of lead time, room type, and special requests
- Visualize key patterns and insights using charts


## üìÅ Dataset Information
The dataset contains hotel booking records with features such as:
- Booking dates
- Hotel type
- Lead time
- Arrival date
- Number of guests
- Special requests
- Cancellation status


## üß† Hypotheses & Assumptions

To guide the exploratory data analysis, the following hypotheses and assumptions were made:

### üìå Hypotheses:
1. **H1:** Cancellations are more likely when the lead time is longer.
2. **H2:** Guests who make special requests are less likely to cancel.
3. **H3:** Booking cancellations are higher during peak months (like July-August).
4. **H4:** Resort hotels have a different booking pattern compared to city hotels.
5. **H5:** Guests with children tend to book longer stays.

### üìå Assumptions:
- The dataset is representative of actual booking trends.
- All variables are correctly recorded and accurately labeled.
- Missing values do not significantly distort overall patterns.
- The "is_canceled" feature accurately reflects actual cancellations.
- No external promotions or events dramatically affected bookings during the time of data collection.

These hypotheses were tested using visualizations and statistical insights throughout the analysis.

```python
plt.figure(figsize = (8,4))
sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
plt.legend(bbox_to_anchor = (1,1))
plt.xlabel('hotel')
plt.ylabel('Number of resarvation')
plt.legend(['Not canceled','is canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
City_hotel = df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# print(city_hotel)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
plt.figure(figsize = (20,8))
plt.title('Average Daily rate of City and Resort Hotel')
plt.plot(resort_hotel.index, resort_hotel['adr'],label = 'Resort hotel')
plt.plot(city_hotel.index, city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
ax1.legend(bbox_to_anchor = (1,1))
plt.xlabel =  'Months'
plt.ylabel = 'Number of resarvation'
plt.legend(['not canceled','is Canceled'])
```

```python
df.head(2)
```

```python
canceled_data = df[df['is_canceled'] == 1]
not_canceled_data = df[df['is_canceled'] == 0]
```

```python
plt.figure(figsize = (15,5))
plt.title('Average Daily Rate',fontsize = 20)
sns.barplot(x = 'month',y = 'adr',data = canceled_data.groupby('month')[['adr']].sum().reset_index(),palette = 'Reds')
plt.show()
```

```python
top_10_country = canceled_data['country'].value_counts()[:10]
```

```python
plt.figure(figsize = (8,8))
plt.pie(top_10_country,autopct = '%.2f',labels = top_10_country.index)
plt.show()
```

```python
cancelled_data_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
cancelled_data_adr.sort_values('reservation_status_date',inplace = True)

Not_cancelled_data = not_canceled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
Not_cancelled_data.sort_values('reservation_status_date',inplace = True)
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = (20))
plt.plot(Not_cancelled_data['reservation_status_date'],Not_cancelled_data['adr'],label = 'Not cancelled')
plt.plot(cancelled_data_adr['reservation_status_date'],cancelled_data_adr['adr'],label = 'Cancelled')
plt.legend()
```

```python
cancelled_data_adr = cancelled_data_adr[(cancelled_data_adr['reservation_status_date'] > '2016') & (cancelled_data_adr['reservation_status_date'] < '2017-09')]
Not_cancelled_data = Not_cancelled_data[(Not_cancelled_data['reservation_status_date'] > '2016') & (Not_cancelled_data['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 20)
plt.plot(Not_cancelled_data['reservation_status_date'],Not_cancelled_data['adr'],label = 'Not cancelled')
plt.plot(cancelled_data_adr['reservation_status_date'],cancelled_data_adr['adr'],label = 'Cancelled')
plt.legend()
```
